```python
from plot_utils import data_dict
from numpy import mean
import pandas as pd
```

```python
# seeds
seeds = [224, 389, 405, 432, 521, 580, 639, 673, 803, 869]
# envs
envs = ["Acrobot-v1", "CartPole-v1", "MountainCar-v0", "MountainCarContinuous-v0", "Pendulum-v1", "LunarLander-v2"]
# methods
methods = ["binQ", "CAT-RL", "TileCoding"]
```

```python
from numpy import std


data = data_dict(methods, envs, seeds)

l = []
for env in envs:
    for method in methods:
        success_rate = []
        accumulated_reward = []
        for seed in seeds:
            success_rate.append(mean(data[method][env][seed]["success"]))
            accumulated_reward.append(sum(data[method][env][seed]["accumulated reward"]))
        
        avg_success_rate = mean(success_rate)
        avg_success_rate_std = std(success_rate)
        avg_accumulated_reward = mean(accumulated_reward)
        avg_accumulated_reward_std = std(accumulated_reward)

        l.append({"method": method, "env": env, "success rate": avg_success_rate, "success rate sts": avg_success_rate_std, "accumulated reward": avg_accumulated_reward, "accumulated reward std": avg_accumulated_reward_std})

df = pd.DataFrame(l).sort_values(by=["method", "env"])

df
```

```python
data = data_dict(methods, envs, seeds, folder="results-after-train")

l = []
for env in envs:
    for method in methods:
        success_rate = []
        accumulated_reward = []
        for seed in seeds:
            success_rate.append(mean(data[method][env][seed]["success"]))
            accumulated_reward.append(sum(data[method][env][seed]["accumulated reward"]))
        avg_success_rate = mean(success_rate)
        avg_accumulated_reward = mean(accumulated_reward)

        l.append({"method": method, "env": env, "success rate": avg_success_rate, "accumulated reward": avg_accumulated_reward})

df = pd.DataFrame(l).sort_values(by=["method", "env"])
df
```
